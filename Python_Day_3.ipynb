```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

ps = pd.read_csv('/content/gdrive/My Drive/CRT_PYTHON_22BK1A7248/DEC NEW DATASET.csv')

print("Missing Values:\n", ps.isnull().sum())

numeric_ps = ps.select_dtypes(include=['int64', 'float64'])

correlation_matrix = numeric_ps.corr()

print("\nCorrelation Matrix:\n", correlation_matrix)

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Correlation Matrix of Shopping Data")
plt.tight_layout()
plt.show()
```
